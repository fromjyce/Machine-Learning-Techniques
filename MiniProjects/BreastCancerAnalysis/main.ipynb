```python
import pandas as pd

# Load the dataset from UCI ML Repository (or local CSV)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
columns = ["ID", "Diagnosis"] + [f"Feature_{i}" for i in range(1, 31)]
df = pd.read_csv(url, header=None, names=columns)

# Drop ID column (not useful for ML)
df.drop(columns=["ID"], inplace=True)

# Convert diagnosis to numerical (M = 1, B = 0)
df["Diagnosis"] = df["Diagnosis"].map({"M": 1, "B": 0})

# Display dataset info
df.info()
df.head()
```

```python
# Basic summary of dataset
print(df.describe())

# Count of each class
print(df["Diagnosis"].value_counts())

# Check for missing values
print(df.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x=df["Diagnosis"])
plt.title("Benign vs Malignant Cases")
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=False, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df, hue="Diagnosis", vars=["Feature_1", "Feature_2", "Feature_3", "Feature_4"])
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x="Diagnosis", y="Feature_1", data=df)
plt.title("Distribution of Feature_1 (Mean Radius) by Diagnosis")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

# Load Dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
columns = ["ID", "Diagnosis"] + [f"Feature_{i}" for i in range(1, 31)]
df = pd.read_csv(url, header=None, names=columns)
df.drop(columns=["ID"], inplace=True)
df["Diagnosis"] = df["Diagnosis"].map({"M": 1, "B": 0})

# Set visual style
sns.set_style("whitegrid")
plt.rcParams.update({'font.size': 12, 'axes.labelsize': 12, 'axes.titlesize': 14})

### 1ï¸âƒ£ Class Distribution (Benign vs Malignant)
plt.figure(figsize=(6, 4))
sns.countplot(x=df["Diagnosis"], palette=["green", "red"])
plt.xticks([0, 1], ["Benign", "Malignant"])
plt.title("Distribution of Diagnosis Classes")
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.savefig("class_distribution.png", dpi=300, bbox_inches="tight")  # Save for PPT
plt.show()

### 2ï¸âƒ£ Feature Distributions (Histogram)
df.drop(columns=["Diagnosis"]).hist(figsize=(15, 12), bins=20, color="skyblue", edgecolor="black")
plt.suptitle("Feature Distributions", fontsize=16)
plt.savefig("feature_distributions.png", dpi=300, bbox_inches="tight")
plt.show()

### 3ï¸âƒ£ Correlation Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=False, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.savefig("correlation_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()

### 4ï¸âƒ£ Feature Importance (Using Random Forest)
X, y = df.drop(columns=["Diagnosis"]), df["Diagnosis"]
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)
feat_importances = pd.Series(rf.feature_importances_, index=X.columns)
plt.figure(figsize=(8, 6))
feat_importances.nlargest(10).plot(kind="barh", color="royalblue")
plt.title("Top 10 Important Features")
plt.savefig("feature_importance.png", dpi=300, bbox_inches="tight")
plt.show()

### 5ï¸âƒ£ Boxplots to Show Feature Distribution by Class
plt.figure(figsize=(8, 5))
sns.boxplot(x="Diagnosis", y="Feature_1", data=df, palette=["green", "red"])
plt.xticks([0, 1], ["Benign", "Malignant"])
plt.title("Feature_1 (Radius Mean) vs Diagnosis")
plt.savefig("boxplot_feature1.png", dpi=300, bbox_inches="tight")
plt.show()

### 6ï¸âƒ£ Pairplot to Show Feature Relationships
sns.pairplot(df, hue="Diagnosis", vars=["Feature_1", "Feature_2", "Feature_3", "Feature_4"], palette=["green", "red"])
plt.savefig("pairplot.png", dpi=300, bbox_inches="tight")
plt.show()
```

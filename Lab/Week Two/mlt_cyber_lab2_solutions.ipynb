Load iris dataset using scikit learn.



```python
# Experiment 1
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np

iris = load_iris()
X = np.array(iris["data"])
Y = np.array(iris["target"])
category_names = iris["target_names"]
feature_names = iris["feature_names"]
```

```python

print(category_names)
print(feature_names)
```

```python
# Description of the IRIS Dataset
print("Feature names:", feature_names)
print("Target names:", category_names)
print("Number of samples:", len(X))
print("Number of features:", len(feature_names))
print("Target labels:", category_names)
```

```python
for i, feature_name in enumerate(feature_names):
    feature_values = X[:, i]
    print(f"\nStatistics for {feature_name}:")
    print("  Mean:", feature_values.mean())
    print("  Standard Deviation:", feature_values.std())
    print("  Min:", feature_values.min())
    print("  Max:", feature_values.max())
```

```python
print("\nFirst 5 samples:")
for i in range(5):
    sample = X[i]
    target = Y[i]
    print(f"Sample {i + 1}: {sample} - Target: {iris.target_names[target]}")
```

## Covariance matrix
Create a numpy array with 3 features land size, price, number of bedrooms. the size of the array will be 3xN. N is the number of samples.

An example of how to find the covariance matrix is given below:

```python
import numpy as np
# Here number of samples is 5 and 3 features
x1 = [4.01, -10.2,  -4.3, 8.4, 3.1] # feature 1
x2 = [2.3,  11.1,  6.12, -0.123, 0.97] # feature 2
x3 = [9.0, 7.1, -7.1, -9.87, 6.23] # feature 3
X = np.stack((x1, x2, x3), axis=0)
np.cov(X)
# This gives the correlation between features
# A(2,3) gives the relation between feature 2 and 3
# A(2,2) gives the variance in feature 2
# Browse through and find out why this matrix is much needed in machine learning
```

# Why Covariance matrix is much needed in machine learning

To understand the relationships between variables in a dataset. It helps us identify the strength and direction of the linear relationship between pairs of variables.

```python
#Create a numpy array with 3 features land size, price, number of bedrooms. the size of the array will be 3xN. N is the number of samples.
import numpy as np
N = np.random.randint(5, 15)
land_size = np.random.rand(N) * 1000
price = np.random.rand(N) * 1000000
num_bedrooms = np.random.randint(1, 6, size=N)

features_array = np.array([land_size, price, num_bedrooms])
features_array = features_array.T

print(features_array)
print("-------")
covariance_matrix = np.cov(features_array, rowvar=False)

print(covariance_matrix)

```

### Find the covariance matrix for entire real_estate database.

```python
covariance_matrix = df.cov()
print(covariance_matrix)
```

## Basic linear algebra
- Find the eigen values of the covariance matrix of the entire dataset (use eigen value decomposition)
- Find the singular values of the covariance matrix of the entire dataset

```python
#Find the eigen values of the covariance matrix of the entire dataset (use eigen value decomposition)
eigenvalues, _ = np.linalg.eig(covariance_matrix)

print(eigenvalues)
```

```python
#Find the singular values of the covariance matrix of the entire dataset
_, singular_values, _ = np.linalg.svd(covariance_matrix)

print(singular_values)
```

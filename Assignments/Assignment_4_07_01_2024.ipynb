Assignment:

Scatterplot, Boxplot, Heatmap, Histogram, Distribution, Distance Histogram

Dataset Link: https://data.world/carlvlewis/interdependence-ngos-by-funding-reach

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("fund-data.csv")

print(dataset)

print("-------------------------------------")

print(dataset.info())

print("----------------------------------------")

print(dataset.describe())
```

**Scatter Plot**

```python
numerical_columns = dataset.select_dtypes(include='number').columns
print(numerical_columns)
```

```python
import random

selected_features = random.sample(list(numerical_columns), 2)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=selected_features[0], y=selected_features[1], data=dataset)
plt.title(f'Scatter Plot: {selected_features[0]} vs. {selected_features[1]}')
plt.xlabel(selected_features[0])
plt.ylabel(selected_features[1])
plt.show()
```

**BoxPlot**

```python
dataset['Funding (2015)'] = pd.to_numeric(dataset['Funding (2015)'], errors='coerce')
dataset['Funding (2016)'] = pd.to_numeric(dataset['Funding (2016)'], errors='coerce')
dataset['Funding (2017)'] = pd.to_numeric(dataset['Funding (2017)'], errors='coerce')

plt.figure(figsize=(12, 6))
sns.boxplot(data=dataset[['Funding (2015)', 'Funding (2016)', 'Funding (2017)']])
plt.title('Boxplot: Funding Distribution Over Years')
plt.show()
```

**HeatMap**

```python
correlation_matrix = dataset.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap: Correlation Matrix')
plt.show()
```

**Distribution Graph**

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(dataset['Est. People Impacted'], fill=True)
plt.title('Distribution Plot: Estimated People Impacted')
plt.xlabel('Estimated People Impacted')
plt.ylabel('Density')
plt.show()
```

**Histogram Graph**

```python
plt.figure(figsize=(10, 6))
sns.histplot(dataset['Number of Employees'], bins=20, kde=True)
plt.title('Histogram: Number of Employees')
plt.xlabel('Number of Employees')
plt.ylabel('Frequency')
plt.show()
```

**Distance Histogram**

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=dataset, x='Twitter Followers', y='FB Likes', bins=20)
plt.title('Distance Histogram: Twitter Followers vs. FB Likes')
plt.xlabel('Twitter Followers')
plt.ylabel('FB Likes')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class PolynomialRegression:
    def __init__(self, degree, learning_rate=0.01, n_iterations=1000):
        self.degree = degree
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None

    def fit(self, X, y):
        X_poly = self._polynomial_features(X, self.degree)
        self.weights = np.zeros(X_poly.shape[1])

        for _ in range(self.n_iterations):
            predictions = np.dot(X_poly, self.weights)
            error = predictions - y
            gradient = np.dot(X_poly.T, error) / len(X_poly)
            self.weights -= self.learning_rate * gradient

    def predict(self, X):
        X_poly = self._polynomial_features(X, self.degree)
        return np.dot(X_poly, self.weights)

    def _polynomial_features(self, X, degree):
        X_poly = np.ones((len(X), 1))
        for d in range(1, degree + 1):
            X_poly = np.concatenate((X_poly, X ** d), axis=1)
        return X_poly
```

```python
# Generate synthetic data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 3 * X.squeeze()**2 + 5 * X.squeeze() + np.random.randn(100)

# Fit polynomial regression model
degree = 2
model = PolynomialRegression(degree=degree)
model.fit(X, y)

# Make predictions
X_new = np.linspace(0, 2, 100).reshape(-1, 1)
y_pred = model.predict(X_new)

# Plot the results
plt.scatter(X, y, label='Data')
plt.plot(X_new, y_pred, color='red', label=f'Polynomial Regression (Degree {degree})')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
